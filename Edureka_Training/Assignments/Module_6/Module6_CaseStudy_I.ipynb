```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('prisoners.csv')
#Load the dataset “prisoners.csv” using pandas and display the first and last five rows in the dataset.
print('First 5 rows: \n')
print(df.head(5))
print('Last 5 rows: \n')
print(df.tail(5))
```

```python
#Use describe method in pandas and find out the number of columns. Can you say something about those rows who have zero inmates?
df.describe()
```

```python
#Create a new column -’total_benefitted’ that is a sum of inmates benefitted through all modes.
df['total_benefitted'] = df['No. of Inmates benefitted by Elementary Education']+df[
    'No. of Inmates benefitted by Adult Education']+df['No. of Inmates benefitted by Higher Education']+df[
    'No. of Inmates benefitted by Computer Course']

df
```

```python
#Create a new row - “totals” that is the sum of all inmates benefitted through each mode across all states.
df = df.append(df.sum(), ignore_index=True)
df
df.at[35, "STATE/UT"] = "totals"
df.at[35, "YEAR"] = "0"
df
```

```python
#dropping the total row
df.drop(35, inplace=True)
```

```python
#Make a bar plot with each state name on the x -axis and their total benefitted inmates as their bar heights. Which state has the maximum number of beneficiaries?

plt.bar(x= df['STATE/UT'],height=df.total_benefitted)
plt.setp(plt.gca().get_xticklabels(), rotation=90,horizontalalignment='right')
plt.show()
```

```python
#Make a pie chart that depicts the ratio among different modes of benefits.
plt.pie(df.total_benefitted, labels = df['STATE/UT'], autopct='%1.1f%%', shadow=True,startangle=140)
plt.show()
```
