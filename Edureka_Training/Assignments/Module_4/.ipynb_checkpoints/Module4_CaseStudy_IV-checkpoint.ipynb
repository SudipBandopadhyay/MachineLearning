```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('BigMartSalesData.csv')
df.head()
```

# Plot Total Sales Per Month for Year 2011. How the total sales have increased over months in Year 2011. Which month has lowest Sales?

```python
df_2011 = df[df.Year == 2011]
df_2011.head()
```

```python
df_2011_sales= df_2011.groupby('Month').agg({'Amount':np.sum})
print(df_2011_sales)
df_2011_sales.plot()
```

Month Feb has the lowest sales in 2011

# Plot Total Sales Per Month for Year 2011 as Bar Chart. Is Bar Chart Better to visualize than Simple Plot?

```python
df_2011_sales.plot(kind='bar')
```

Bar chart has much better visibility than line chart to check for the sales each month. Line charts are better to check for trends.

# Plot Pie Chart for Year 2011 Country Wise. Which Country contributes highest towards sales?

```python
df_2011_country= df_2011.groupby('Country').agg({'Amount':np.sum})
df_2011_country
```

```python
plt.figure(figsize=(7,7))
plt.pie(df_2011_country.Amount, labels=df_2011_country.index,autopct='%1.1f%%',shadow=True, startangle=90)
plt.show()
```

# Plot Scatter Plot for the invoice amounts and see the concentration of amount. In which range most of the invoice amounts are concentrated

```python
df_2011_invoice= df_2011.groupby('InvoiceNo').agg({'Amount':np.sum})
df_2011_invoice
```

```python
plt.scatter(df_2011_invoice.Amount,df_2011_invoice.Amount,color='blue')
```
