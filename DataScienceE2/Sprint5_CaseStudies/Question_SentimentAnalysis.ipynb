### Perform Sentiment Analysis of the given data set and store it as the 'sentiment.csv' and the column names should be 'content', 'sentiment' and 'value'

- Column 'content' has the text as given in the dataset

### Follow the Instructions given below

- Read the data from the CSV file
- Conver the content to lower case and remove the numbers, special charackers
- Remove the stop words from the content
- Create two columns 'sentiment' and 'value'
- Predict the sentiment for the given data and store it to the respective columns
- Once you done with the above steps save the dataframe as 'sentiment.csv' 

```python
#Write your code here and add more cells as needed
sentencelist = review.content.tolist()
```

```python
stopword = stopwords.words('english')
sid_obj =SentimentIntensityAnalyzer()
sentiment=[]
for i in range(len(sentencelist)):
    sentence=sentencelist[i].lower()
    words = [word for word in sentence.split() if word.lower() not in stopword]
    new_text = " ".join(words)
    new_text = re.sub('[^A-Za-z]+', ' ', new_text)
    sentiment_dict = sid_obj.polarity_scores(new_text)
    print(sentence)
    print(new_text)
    print(str(sentiment_dict))
    if sentiment_dict['compound'] >= 0.05 :
        sentiment.append(['Positive',1])
        #print("Positive")
 
    else:
        sentiment.append(['Negative',0])
        #print("Negative")
```

```python
sentiment_df=pd.DataFrame(data=sentiment,columns=['sentiment','value'])
sentiment_df
```

```python
review_df = review.copy()
review_df[['sentiment','value']] = sentiment_df[['sentiment','value']].copy()
review_df
```

```python
review_df.to_csv('sentiment.csv')
```
